## How to import Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Matplotlib Types

**i) Basic Line Plot :-** Create a simple line ploat using Matplotlib. It involves defining data points using the plt.plot() function.

```python
x = np.linspace(0,10,50)
y = np.sin(x)
plt.plot(x,y,label=' Sine Wave ')
plt.xlabel(' X-axis ')
plt.ylabel(' Y-axis ')
plt.title(' Simple Line Plot ')
plt.legend()
plt.show()
```

* **x = np.linspace(0, 10, 50) :-** The linspace() function of NumPy is used to create the variable x with 50 equally spaced values between 0 and 10.
* **y = np.sin(x) :-** The sine function is applied to the values of x to generate the corresponding sine wave values, which are stored in y.
* **plt.plot(x, y, label='Sine Wave') :-** This line plots the x and y values on the graph and assigns the label "Sine Wave" to the line.
* **plt.xlabel('X-axis') :-** This sets the label for the x-axis of the graph.
* **plt.ylabel('Y-axis') :-** This sets the label for the y-axis of the graph.
* **plt.title('Simple Line Plot') :-** This sets the title of the graph, describing the plotted data.
* **plt.legend() :-** This displays the legend on the graph showing the label "Sine Wave".
* **plt.show() :-** This displays the final line plot on the screen.

**ii) Scatter Plot :-** Scatter Plots are useful for visualizing relationship between two variables.

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y,color='b',marker='o')
plt.xlabel(' X-axis ')
plt.ylabel(' Y-axis ')
plt.title(' Scatter Plot ')
plt.show()
```

* **x = np.random.rand(50) :-** This line generates 50 random values between 0 and 1 and stores them in the variable x.
* **y = np.random.rand(50) :-** This line generates 50 random values between 0 and 1 and stores them in the variable y.
* **plt.scatter(x, y, color='b', marker='o') :-** This creates a scatter plot using the x and y values.The points are shown in blue color (b) and circle (o) shape.
* **plt.xlabel('X-axis') :-** This sets the label for the x-axis.
* **plt.ylabel('Y-axis') :-** This sets the label for the y-axis.
* **plt.title('Scatter Plot') :-** This sets the title of the graph.
* **plt.show() :-** This displays the scatter plot on the screen.

**iii) Bar Charts :-** A bar charts represents data with rectangular bars.

```python
categories = ['A', 'B', 'C', 'D']
values =[3, 7, 1, 8]
plt.bar(categories,values,color='Blue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')
plt.show()
```

* **categories = ['A', 'B', 'C', 'D'] :-** This line creates a list named categories that contains the category labels for the bar chart.
* **values = [3, 7, 1, 8] :-** This line creates a list named values that stores the numerical values corresponding to each category.
* **plt.bar(categories, values, color='Blue') :-** This function creates a bar chart where each category is represented by a bar.The bars are displayed in blue color.
* **plt.xlabel('Categories') :-** This sets the label for the x-axis.
* **plt.ylabel('Values') :-** This sets the label for the y-axis.
* **plt.title('Bar Chart') :-** This sets the title of the bar chart.
* **plt.show() :-** This displays the bar chart on the screen.

**iv) Histogram :-** Histograms represent the distribution of numerical data.

```python
data = np.random.randn(1000)
plt.hist(data,bins=30,color='green',edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

* **data = np.random.randn(1000) :-** This line generates 1000 random values from a normal (Gaussian) distribution and stores them in the variable data.
* **plt.hist(data, bins=30, color='green', edgecolor='black') :-** This function creates a histogram of the data.
     * bins=30 divides the data into 30 intervals (bars)
     * color='green' sets the color of the bars
     * edgecolor='black' adds a black border to each bar.
* **plt.xlabel('Value') :-** This sets the label for the x-axis, representing the data values.
* **plt.ylabel('Frequency') :-** This sets the label for the y-axis, representing how often the values occur.
* **plt.title('Histogram') :-** This sets the title of the histogram
* **plt.show() :-** This displays the histogram on the screen.

**v) Pie Chart :-** Pie  chat usrd to show percentages or proport


```python
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 30, 35]
plt.pie(sizes, labels=labels)
plt.title("Pie Chart")
plt.show()
```

* **labels = ['Python', 'Java', 'C++', 'JavaScript'] :-** This line creates a list called labels that contains the names of categories shown in the pie chart.
* **sizes = [40, 25, 30, 35] :-** This line creates a list called sizes that represents the values of each category.
* **plt.pie(sizes, labels=labels) :-** This function creates a pie chart using the values from sizes.Each slice of the pie is labeled using the values from labels.
* **plt.title("Pie Chart") :-** This sets the title of the pie chart.
* **plt.show() :-** This displays the pie chart on the screen.

**vi) Subplots :-** Matplotlib allows creating plots in one figure using subplots().

```python
fig,ax=plt.subplots(1,2,figsize=(10,4))
ax[0].plot(x,y,color='Blue',linestyle='--')
ax[0].set_title('Sine Wave')
ax[1].bar(categories,values,color='Red')
ax[1].set_title('Bar Chart')
plt.show()
```

* **fig, ax = plt.subplots(1, 2, figsize=(10, 4)) :-** This line creates a figure with 1 row and 2 columns of subplots.figsize=(10,4) sets the width and height of the figure.
* **ax[0].plot(x, y, color='Blue', linestyle='--') :-** This plots a line graph on the first subplot using x and y values.The line is shown in blue color with a dashed style.
* **ax[0].set_title('Sine Wave') :-** This sets the title of the first subplot as Sine Wave.
* **ax[1].bar(categories, values, color='Red') :-** This creates a bar chart on the second subplot using categories and values.The bars are displayed in red color.
* **ax[1].set_title('Bar Chart') :-** This sets the title of the second subplot as Bar Chart.
* **plt.show() :-** This displays both plots together in one figure

**vii) Box Plot :-** A Box plot(also called a box-and-whisker plot) is a statistical chart that shows the distribution of a dataset.

```python
data=[np.random.normal(0,1,100),
      np.random.normal(2,1.5,100),
      np.random.normal(-1,0.5,100)]
plt.boxplot(data)
plt.title("Basic Box Plot")
plt.xlabel("Dataset")
plt.ylabel("Value")
plt.show()
```

* **data = [np.random.normal(0, 1, 100), np.random.normal(2, 1.5, 100), np.random.normal(-1, 0.5, 100)] :-** This line creates a list named data that contains three datasets.
    * The first dataset has a mean of 0 and standard deviation of 1
    * The second dataset has a mean of 2 and standard deviation of 1.5
    * The third dataset has a mean of -1 and standard deviation of 0.5
    * Each dataset contains 100 random values.
* **plt.boxplot(data) :-** This function creates a box plot for each dataset.
   * shows the median, quartiles, and spread of the data.
* **plt.title("Basic Box Plot") :-** This sets the title of the box plot.
* **plt.xlabel("Dataset") :-** This sets the label for the x-axis, representing different datasets.
* **plt.ylabel("Value") :-** This sets the label for the y-axis, representing the data values.
* **plt.show() :-** This displays the box plot on the screen.

**viii) Density Plot :-** A density plot (also called a KDE plot- Kernal Density Estimate) is a smooth curve that shows the probability distribution of a continuous variable. It is similar to a smoothed histogram.

```python
pulse = [70, 72, 75, 78, 80, 76, 74, 82, 79, 73]
plt.hist(pulse, density=True)
plt.title("Density Plot of Pulse")
plt.xlabel("Pulse")
plt.ylabel("Density")
plt.show()
```

* **pulse = [70, 72, 75, 78, 80, 76, 74, 82, 79, 73] :-** This line creates a list named pulse that contains pulse rate values.
* **plt.hist(pulse, density=True) :-** This function creates a histogram of the pulse data.
    * Setting density=True converts the histogram into a density plot, where the total area under the bars is equal to 1.
* **plt.title("Density Plot of Pulse") :-** This sets the title of the graph.
* **plt.xlabel("Pulse") :-** This sets the label for the x-axis, representing pulse values.
* **plt.ylabel("Density") :-** This sets the label for the y-axis, representing the density of the data.
* **plt.show() :-** This displays the density plot on the screen.
